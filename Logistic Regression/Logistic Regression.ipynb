```python
import numpy as np;
import pandas as pd;
from matplotlib import pyplot as plt;

df = pd.read_csv(r"C:\Users\Raider\Downloads\heart.csv");
df
```

```python
X = df.drop(['TenYearCHD'], axis = 1)
Y = df.TenYearCHD

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import r2_score

model = LogisticRegression();
model.fit(X,Y)
y_pred = model.predict(X)

print('Intercept: ', model.intercept_)
print('Coefficient: ', model.coef_)
print('Predicted Values: ', model.predict(X))
print('Confusion matrix: ', metrics.confusion_matrix(Y, y_pred))
print('Mean Absolute Error: ', metrics.mean_absolute_error(Y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(Y, y_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(Y, y_pred)))
print('R2 Score', r2_score(Y, y_pred))
print('Predicted Values Probability: ', model.predict_proba(X))
```
