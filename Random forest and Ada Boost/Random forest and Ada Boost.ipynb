```python
import numpy as np;
import pandas as pd;
from matplotlib import pyplot as plt;

df = pd.read_csv(r"C:\Users\Raider\Downloads\heart.csv");
df
```

```python
X = df.drop(['TenYearCHD'], axis = 1)
Y = df.TenYearCHD

from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.1,random_state=0)
```

```python
#Decision Tree created (for comparison)
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```

```python
#Stumps created (also for comparison)
model = DecisionTreeClassifier(max_depth=1)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50, max_features="auto")
#n_estimators: no. of trees in forest(default: 100)
#max_features = auto implies max feature = sqrt(p_features)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```

```python
model = RandomForestClassifier(n_estimators=10, max_features="log2")
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```

```python
#Too many unnecessary features also reduce score
model = RandomForestClassifier(n_estimators=10, max_features=15)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```

```python
#Ada Boost Classifier
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=50)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("Predicted values for test data:", y_pred)
print("Score for training data:", model.score(x_train,y_train))
print("Score for test data:", model.score(x_test,y_test))
```
