```python
import numpy as np;
import pandas as pd;
from matplotlib import pyplot as plt;

df = pd.read_csv(r"C:\Users\Raider\Downloads\bank_note_data.csv");
print(df)
```

```python
X = df.drop(['Class'], axis = 1)
Y = df.Class

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=1000, activation='relu')
mlp.fit(x_train, y_train)
print('training score:', mlp.score(x_train, y_train))
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score

y_pred = mlp.predict(x_test)
print('Confusion matrix: ', metrics.confusion_matrix(y_test, y_pred))
print('test score:', mlp.score(x_test, y_test))
print('No. of nodes in output: ', mlp.n_outputs_)
print('Error/Activation function:', mlp.out_activation_)
print('Weights from first to final layer):', mlp.coefs_)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=500, activation='logistic')
mlp.fit(x_train, y_train)

from sklearn import metrics
from sklearn.metrics import r2_score

y_pred = mlp.predict(x_test)
print('Confusion matrix: ', metrics.confusion_matrix(y_test, y_pred))
print('training score:', mlp.score(x_train, y_train))
print('test score:', mlp.score(x_test, y_test))
print('No. of nodes in output: ', mlp.n_outputs_)
print('Error/Activation function:', mlp.out_activation_)
print('Weights from first to final layer):', mlp.coefs_)
```
